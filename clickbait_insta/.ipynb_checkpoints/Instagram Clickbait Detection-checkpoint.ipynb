```python
#importing basic packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the data
data0 = pd.read_csv('InstagramData_ICWSM2018_180406.csv')
data0.head()
```

```python
data0=data0.drop(['Likes', 'Followings', 'Followers', 'Comments', 'MediaCounts', 'LocationExistence'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data0['SearchedTag'] = le.fit_transform(data0['SearchedTag'])
```

```python
data0.describe()
```

```python
#Dropping the Domain column
data=data0.drop(['Link','ImageURL', 'Hashtags', 'Captions'], axis=1)
#data = data0.copy()
data
```

```python
data['Logo'] = data['Logo'].fillna((data['Logo'].mean()))
data['ProductOnly'] = data['ProductOnly'].fillna((data['ProductOnly'].mean()))
```

```python
# shuffling the rows in the dataset so that when splitting the train and test set are equally distributed
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
# Sepratating & assigning features and target columns to X & y
y = data['Clickbait']
X = data.drop(['Clickbait'],axis=1)
X.shape, y.shape
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import itertools
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB 
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from mlxtend.classifier import StackingCVClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import roc_curve, auc
from sklearn.neural_network import MLPClassifier

fs = SelectKBest(score_func=f_classif, k=200)
X_selected = fs.fit_transform(X,y)
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, 
                                                    test_size = 0.2, random_state = 12)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

RANDOM_SEED = 42

clf1 = KNeighborsClassifier(n_neighbors=1)
clf3 = GaussianNB()
clf2 = LogisticRegression(C=10.0, max_iter=10000)
clf4 = XGBClassifier(learning_rate=0.4,max_depth=7)
clf5 = MLPClassifier(alpha=0.001, hidden_layer_sizes=([100,100,100]))
clf6 = SVC(kernel='linear', C=1.0, random_state=12, probability=True)
forest = RandomForestClassifier(max_depth=4, max_features='sqrt', n_estimators=200, criterion='entropy')

sclf = StackingCVClassifier(classifiers=[clf1, clf2, clf3, clf4, clf5, clf6], 
                            meta_classifier=forest,
                            random_state=42,
                            cv=5)
classifiers = {"KNN": clf1,
               "LR": clf2,
               "NB": clf3,
               "XGBoost": clf4,
               "MLP" : clf5,
               "SVM" : clf6 ,
               "Stack": sclf}

for key in classifiers:
    classifier = classifiers[key]
    classifier.fit(X_train, y_train)
    classifiers[key] = classifier

results = pd.DataFrame()
for key in classifiers:
    y_pred = classifiers[key].predict_proba(X_test)[:,1]
    results[f"{key}"] = y_pred

results["Target"] = y_test
```
