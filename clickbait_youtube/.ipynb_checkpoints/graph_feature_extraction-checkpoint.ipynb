```python
import face_recognition
import imutils
from imutils import paths
import pickle
import time
import cv2
import os

from imageai.Detection import ObjectDetection
import os

def getObjectList(imagepath):
    detector = ObjectDetection()
    detector.setModelTypeAsRetinaNet()
    detector.setModelPath('/Users/peyamowar/Downloads/clickbait_youtube/resnet50_coco_best_v2.1.0.h5')
    detector.loadModel()
    detections = detector.detectObjectsFromImage(input_image=imagepath, output_image_path=os.path.join(imagepath, "new.jpg"))
    names = []
    for eachObject in detections:
        if (eachObject["percentage_probability"]>50):
            names.append(eachObject["name"])
    return names
```

```python
import os
import cv2
import numpy as np

prototxt_path = '/Users/peyamowar/Downloads/clickbait_youtube/model_data/deploy.prototxt'
caffemodel_path = '/Users/peyamowar/Downloads/clickbait_youtube/model_data/weights.caffemodel'
    
def extractAndSaveFaces(imagepath, code):
    model = cv2.dnn.readNetFromCaffe(prototxt_path, caffemodel_path)
    faces_folder_path = "/Users/peyamowar/Downloads/clickbait_youtube/" + code + "/faces/"
    if not os.path.exists(faces_folder_path):
        #print("New directory created")
        os.makedirs(faces_folder_path)
    count = 0
    image = cv2.imread(imagepath)
    (h, w) = image.shape[:2]
    blob = cv2.dnn.blobFromImage(cv2.resize(image, (300, 300)), 1.0, (300, 300), (104.0, 177.0, 123.0))
    model.setInput(blob)
    detections = model.forward()
    for i in range(0, detections.shape[2]):
        box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
        (startX, startY, endX, endY) = box.astype("int")
        confidence = detections[0, 0, i, 2]
        if (confidence > 0.5):
            count += 1
            frame = image[startY:endY, startX:endX]
            cv2.imwrite(faces_folder_path + str(i) + '_' + "face.jpg", frame)
    #print("Extracted " + str(count) + " faces from all images")
    faces_folder_path = "/Users/peyamowar/Downloads/clickbait_youtube/" + code + "/faces/"
    imagePaths = list(paths.list_images(faces_folder_path))
    knownEncodings = []
    knownNames = []
    # loop over the image paths
    for (i, imagePath) in enumerate(imagePaths):
        # extract the person name from the image path
        name = imagePath.split(os.path.sep)[-2]
        # load the input image and convert it from BGR (OpenCV ordering)
        # to dlib ordering (RGB)
        image = cv2.imread(imagePath)
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        #Use Face_recognition to locate faces
        boxes = face_recognition.face_locations(rgb,model='hog')
        # compute the facial embedding for the face
        encodings = face_recognition.face_encodings(rgb, boxes)
        # loop over the encodings
        for encoding in encodings:
            knownEncodings.append(encoding)
            knownNames.append(name)
    #save emcodings along with their names in dictionary data
    data = {"encodings": knownEncodings, "names": knownNames}
    #use pickle to save data into a file for later use
    f = open("/Users/peyamowar/Downloads/clickbait_youtube/"+code+"/face_enc", "wb")
    f.write(pickle.dumps(data))
    f.close()
    return count

def CountOfDetectedFaces(imagepath, code):
    model = cv2.dnn.readNetFromCaffe(prototxt_path, caffemodel_path)
    count = 0
    image = cv2.imread(imagepath)
    (h, w) = image.shape[:2]
    blob = cv2.dnn.blobFromImage(cv2.resize(image, (300, 300)), 1.0, (300, 300), (104.0, 177.0, 123.0))
    model.setInput(blob)
    detections = model.forward()
    for i in range(0, detections.shape[2]):
        #box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
        #(startX, startY, endX, endY) = box.astype("int")
        confidence = detections[0, 0, i, 2]
        if (confidence > 0.5):
            count += 1
            #frame = image[startY:endY, startX:endX]
    return count
```

```python
import face_recognition
import imutils
from imutils import paths
import pickle
import time
import cv2
import os
 
#get paths of each file in folder named Images
#Images here contains my data(folders of various persons)

def numberofcommonfaces(imgpath, code): 
    cascPathface = os.path.dirname(cv2.__file__) + "/data/haarcascade_frontalface_alt2.xml"
    # load the harcaascade in the cascade classifier
    faceCascade = cv2.CascadeClassifier(cascPathface)
    # load the known faces and embeddings saved in last file
    if os.path.isfile("C:/Users/minij/Downloads/Clickbait detection/Keyframes/"+code+"/face_enc")==False:
        return 0
    try:
        data = pickle.loads(open("/Users/peyamowar/Downloads/clickbait_youtube/"+code+"/face_enc", "rb").read())
        image = cv2.imread(imgpath)
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray,
                                     scaleFactor=1.1,
                                     minNeighbors=5,
                                     minSize=(60, 60),
                                     flags=cv2.CASCADE_SCALE_IMAGE)
        encodings = face_recognition.face_encodings(rgb)
        names = []
        for encoding in encodings:
            matches = face_recognition.compare_faces(data["encodings"],
            encoding)
        #set name =inknown if no encoding matches
            name = "Unknown"
        # check to see if we have found a match
            if True in matches:
                print("hello")
            #Find positions at which we get True and store them
                matchedIdxs = [i for (i, b) in enumerate(matches) if b]
                counts = {}
            # loop over the matched indexes and maintain a count for
            # each recognized face face
                for i in matchedIdxs:
                #Check the names at respective indexes we stored in matchedIdxs
                    name = data["names"][i]
                #increase count for the name we got
                    counts[name] = counts.get(name, 0) + 1
                #set name which has highest count
                    name = max(counts, key=counts.get)
            names.append(name)
    except:
        return 0
    i=0
    for name in names:
        if name!="Unknown":
            i=i+1
    return i
```

```python
def findedgeweight(objects_matched, objects_thumbnail, objects_keyframe, faces_matched, faces_thumbnail, faces_keyframe):
    objectweight = 0
    faceweight = 0
    if (objects_thumbnail + objects_keyframe != 0):
        objectweight = (2 * objects_matched)/(objects_thumbnail + objects_keyframe)
    if (faces_thumbnail + faces_keyframe != 0):
        faceweight = (2 * faces_matched)/(faces_thumbnail + faces_keyframe)
    if (objectweight == 0):
        return faceweight
    if (faceweight == 0):
        return objectweight
    return (objectweight + faceweight)/2
```

```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter('ignore')
data = pd.read_csv('MVD_Dataset_Final.csv')[763:770]
#df = pd.read_csv('MVD_Dataset_Final.csv')
#data = df[df['video_id']=='_MzbLQBeR9Y']
main_df = pd.DataFrame(columns = ['Movie ID','Label','Number of Frames','Edge Weights'])
for ind in data.index: 
    code = data['video_id'][ind]
    label = data['LABEL'][ind]
    directory = "/Users/peyamowar/Downloads/clickbait_youtube/" + code
    if (os.path.exists(directory)==False):
        print ("Video not found: "+code)
        continue
    directorykeyframes = directory + "/keyframes"
    thumbnail = directory + "/" + code + "_thumbnail.jpg"
    objectsinthumbnail=getObjectList(thumbnail)  #list of thumbnail obj
    countfacesinthumbnail=extractAndSaveFaces(thumbnail, code)   #count of thumbnail obj
    facesinkeyframes = []    #count of faces in each keyframe : list
    countobjectsinkeyframes = []    #count of obj in each keyframe: list
    commonobjectsforkeyframes = []  #count of onbj common in thumbnail and each keyframe
    commonfacesforkeyframes = []   #count of faces common in thumbnail and each keyframe
    edgeweights = []
    list_dir = os.listdir(directorykeyframes)
    list_dir = [f.lower() for f in list_dir]
    list_dir = sorted(list_dir)
    for file in list_dir:
        if file != '.ds_store':
            imgpath = directorykeyframes+"/"+file
            numberoffacesinframe = CountOfDetectedFaces(imgpath, code)
            n=0
            if (countfacesinthumbnail==0 or numberoffacesinframe==0):
                n=0
            else:
                n = numberofcommonfaces(imgpath, code) 
            commonfacesforkeyframes.append(n)
            facesinkeyframes.append(numberoffacesinframe)
            objectsinkeyframe = getObjectList(imgpath)
            countobjectsinkeyframes.append(len(objectsinkeyframe))
            reqlen=0
            if (len(objectsinkeyframe)!=0):
                newlist = list(set(objectsinthumbnail).intersection(set(objectsinkeyframe)))
                commonobjectsforkeyframes.append(len(newlist))
                reqlen=len(newlist)
            else:
                commonobjectsforkeyframes.append(0)
            edgeweights.append(findedgeweight(reqlen,len(objectsinthumbnail),len(objectsinkeyframe),n,countfacesinthumbnail,numberoffacesinframe))
    movie_dict = dict({'Movie ID': code, 'Label': label, 'Number of Frames':len(commonobjectsforkeyframes), 'Objects in Thumbnail':objectsinthumbnail, 'Total Objects in every Keyframe':countobjectsinkeyframes, 'Objects matched in every Keyframe':commonobjectsforkeyframes, 'Total Faces in Thumbnail':countfacesinthumbnail, 'Total Faces in each Keyframe':facesinkeyframes, 'Faces Matched in each Keyframe':commonfacesforkeyframes})
    #video_df = pd.DataFrame([[code, label, len(commonobjectsforkeyframes), edgeweights]], columns = ['Movie ID','Label','Number of Frames','Edge Weights'])
    main_df.loc[len(main_df.index)] = [code, label, len(commonobjectsforkeyframes), edgeweights]
    dict_file_name = directory + "/" + code + "_dict.txt"
    f = open(dict_file_name,"w")
    f.write(str(movie_dict))
    f.close()
    print("Feature Extraction completed for Video: " + code)
main_df
```
