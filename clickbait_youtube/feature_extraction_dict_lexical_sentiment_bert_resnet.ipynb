```python
#from google_trans_new import google_translator  
from emoji import UNICODE_EMOJI
import string
#translator = google_translator() 

if_numbers_present = list()
if_questions_present = list()
counts_of_emojis = list()
relative_freqs_of_caps = list()
counts_of_punct = list()

for index, row in data.iterrows():
    title = row['Translated Title']
    #title = translator.translate(titlex,lang_tgt='en')
    
    if_number_present = 0
    if_question_present = 0
    count_of_emojis = 0
    relative_freq_of_caps = 0
    count_of_punct = 0
    
    if any(char.isdigit() for char in title):
        if_number_present = 1
    else:
        if_number_present = 0
    if_numbers_present.append(if_number_present)
        
    if '?' in title:
        if_question_present = 1
    else:
        if_question_present = 0
    if_questions_present.append(if_question_present)
        
    for emoji in UNICODE_EMOJI:
        if emoji in title:
            count_of_emojis = count_of_emojis + 1
    counts_of_emojis.append(count_of_emojis)
        
    count_caps = 0
    count_lows = 0
    for i in title:
        if(i.islower()):
            count_lows = count_lows + 1
        elif(i.isupper()):
            count_caps = count_caps + 1
    if count_caps + count_lows == 0:
        relative_freq_of_caps = 0
    else:
        relative_freq_of_caps = count_caps/(count_caps + count_lows)
    relative_freqs_of_caps.append(relative_freq_of_caps)
        
    count = lambda l1,l2: sum([1 for x in l1 if x in l2])    
    count_of_punct = count(title,set(string.punctuation)) 
    counts_of_punct.append(count_of_punct)

dict1 = {'If Number Present' : if_numbers_present, 'If a Question' : if_questions_present, 'Count of Emojis' : counts_of_emojis, 'Relative Freq of Caps' : relative_freqs_of_caps, 'Count of Punctuation Marks' : counts_of_punct}
newdf1 = pd.DataFrame(dict1)
newdf1.head()
```

```python
newdf1.to_csv('lexical_for_bollybait.csv')
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 
sid_obj = SentimentIntensityAnalyzer() 

positive_sentiment = list()
negative_sentiment = list()
neutral_sentiment = list()
compounded_sentiment = list()

for index, row in data.iterrows():
    title = row['Translated Title'].lower()
    #title = translator.translate(titlex,lang_tgt='en') 
    sentiment_dict = sid_obj.polarity_scores(title) 
    negative_sentiment.append(sentiment_dict['neg'])
    positive_sentiment.append(sentiment_dict['pos'])
    neutral_sentiment.append(sentiment_dict['neu'])
    compounded_sentiment.append(sentiment_dict['compound'])
    
dict2 = {'Positive Sentiment' : positive_sentiment, 'Negative Sentiment' : negative_sentiment, 'Neutral Sentiment' : neutral_sentiment, 'Compounded Sentiment' : compounded_sentiment}
newdf2 = pd.DataFrame(dict2)
newdf2.head()
```

```python
import pandas as pd
import numpy as np
import torch
from transformers import BertTokenizer, BertModel
import matplotlib.pyplot as plt
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased',output_hidden_states = True)
```

```python
#from google_trans_new import google_translator 
import string
#translator = google_translator() 
mat = np.empty((0,768), int)
i = 0
for index, row in data.iterrows():
    i = i+1
    print(i)
    title = row['Translated Title']
    #title = translator.translate(titlex,lang_tgt='en')
    title = title.translate(str.maketrans('', '', string.punctuation))
    text = title.strip("[]")
    marked_text = "[CLS] " + text + " [SEP]"
    tokenized_text = tokenizer.tokenize(marked_text)
    indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
    segments_ids = [1] * len(tokenized_text)
    tokens_tensor = torch.tensor([indexed_tokens])
    segments_tensors = torch.tensor([segments_ids])
    with torch.no_grad():
        outputs = model(tokens_tensor, segments_tensors)
        hidden_states = outputs[2]
    token_vecs = hidden_states[-2][0]
    sentence_embedding = torch.mean(token_vecs, dim = 0)
    rowx = np.reshape(sentence_embedding,(1, 768))
    mat = np.append(mat, rowx, axis = 0)
```

```python
import pandas as pd
df_bert = pd.DataFrame(mat)
df_bert.to_csv('bert_for_bollybait.csv')
```
