```python
import numpy as np
import pandas as pd
data = pd.read_csv('mvd_missing.csv')[:165]
data
```

```python
from pytube import YouTube
import urllib.request
import videokf as vf
for ind in data.index: 
    #print(data['Title'][ind])
    print(ind)
    name_tag = data['Video ID'][ind]
    #download youtube video
    youtube_video_url = "https://www.youtube.com/watch?v=" + name_tag
    folder_path = "/Users/peyamowar/Downloads/clickbait_youtube/mvd_missing/" + name_tag + "/"
    video_name = name_tag
    try:
        yt_obj = YouTube(youtube_video_url)
        filters = yt_obj.streams.filter(progressive = True, file_extension = 'mp4')
        filters.get_highest_resolution().download(output_path = folder_path, filename = video_name)
        print('Video Downloaded Successfully')
    except Exception as e:
        print("Error for Video: " + name_tag)
        print(e)
        continue
    #download video thumbnail
    thumbnail_path = "https://img.youtube.com/vi/" + name_tag + "/default.jpg"
    thumbnail_name = folder_path + name_tag + "_thumbnail.jpg"
    try:
        urllib.request.urlretrieve(thumbnail_path, thumbnail_name)
    except Exception as e:
        print(e)
    #extract video keyframes
    keyframe_path = folder_path + name_tag + ".mp4"
    try:
        vf.extract_keyframes(video_file = keyframe_path)
    except Exception as e:
        print(e)
```

```python
#Python 2.x program to transcribe an Audio file 
import speech_recognition as sr 
from google_trans_new import google_translator  
translator = google_translator()  

AUDIO_FILE = ("") 
  
# use the audio file as the audio source 
  
r = sr.Recognizer() 
  
with sr.AudioFile(AUDIO_FILE) as source: 
    #reads the audio file. Here we use record instead of 
    #listen 
    r.adjust_for_ambient_noise(source, duration=0.5)
    audio1 = r.record(source, duration=15)   
    audio2 = r.record(source, duration=15)
    audio3= r.record(source, duration=15)

try: 
    s=r.recognize_google(audio1)
    text1 = s
    text1 = translator.translate(text1,lang_tgt='en')  
    print(text1)
    t=r.recognize_google(audio2)
    text2 = t
    text2 = translator.translate(text2,lang_tgt='en')  
    print(text2)
    f=r.recognize_google(audio3)
    text3 = f
    text3 = translator.translate(text3,lang_tgt='en')  
    print(text3)
  
except sr.UnknownValueError: 
    print("Google Speech Recognition could not understand audio") 
```
